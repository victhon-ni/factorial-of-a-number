```python
num_str=input("enter a number to find its factorial:")
#check if the input is empty
if num_str.strip()=="":
    print("no input found.please enter a number.")
else:
    if not num_str.lstrip('_').isdigit():    #check if input is valid int (not letters or symbol)
        print("please enter a valid integer.")
    else:
        num=int(num_str)   #convert string to integer
        if num<0:     #factorial of negative numbers doesn't exist
            print("factorial does not exist for negative numbers.")
        else:
            factorial=1    #start with 1
            for i in range (1,num+1):    #loop to multiply numbers from 1 to n
                factorial*=i     
            print("factorial of ",num,"is:",factorial)
```
